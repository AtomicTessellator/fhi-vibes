```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dos1 = np.loadtxt('total_dos_111.dat')[:, 1]
dos2 = np.loadtxt('total_dos_222.dat')[:, 1]
dos3 = np.loadtxt('total_dos_333.dat')[:, 1]
dos5 = np.loadtxt('total_dos_555.dat')[:, 1]
```

```python
# normiere
e1 = dos1 / np.linalg.norm(dos1)
e2 = dos2 / np.linalg.norm(dos2)
e3 = dos3 / np.linalg.norm(dos3)
e5 = dos5 / np.linalg.norm(dos5)
```

```python
fix, axs = plt.subplots(nrows=4)

axs[0].plot(e1)
axs[1].plot(e2)
axs[2].plot(e3)
axs[3].plot(e5)

plt.show()
```

```python
# scalar product
for e in (e1, e2, e3):
    print(f'{e5 @ e * 100:.2f} %')
```
